```python
# import pandas, matplotlib, numpy, and seaborn

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import random
import random

#Import stats from scipy
from scipy import stats

# import math
import math
```

# For Developed Countries

## Get GDP Data

```python
df_gdp_developed = pd.read_csv('gdp_developed.csv')
df_gdp_developed['year'] = pd.to_datetime(df_gdp_developed['year'], format='%Y').dt.year
df_gdp_developed = df_gdp_developed[df_gdp_developed['year'] >= 2012]
df_gdp_developed = df_gdp_developed.reset_index(drop=True)

# Drop the columns that contains NaN values
df_gdp_developed = df_gdp_developed.dropna(axis=1)
```

## Get Export Data

```python
df_export_developed = pd.read_csv('export_developed.csv')
df_export_developed['year'] = pd.to_datetime(df_export_developed['year'], format='%Y').dt.year
df_export_developed = df_export_developed[df_export_developed['year'] >= 2012]
df_export_developed = df_export_developed.reset_index(drop=True)

# Drop the columns that contains NaN values
df_export_developed = df_export_developed.dropna(axis=1)
```

## Get Import Data

```python
df_import_developed = pd.read_csv('import_developed.csv')
df_import_developed['year'] = pd.to_datetime(df_import_developed['year'], format='%Y').dt.year
df_import_developed = df_import_developed[df_import_developed['year'] >= 2012]
df_import_developed = df_import_developed.reset_index(drop=True)

# Drop the columns that contains NaN values
df_import_developed = df_import_developed.dropna(axis=1)
```

## Finalizing the data

```python
gdp_developed_columns = df_gdp_developed.columns
export_developed_columns = df_export_developed.columns
import_developed_columns = df_import_developed.columns

intesection_columns = set(gdp_developed_columns).intersection(set(export_developed_columns)).intersection(set(import_developed_columns))

# Remove entry 'year' from the list
intesection_columns.remove('year')

# Convert the intersection columns to a list
intesection_columns = list(intesection_columns)

# Randomly select the 40 strings from the list
intesection_columns = random.sample(intesection_columns, 40)

# Append 'year' to the list
intesection_columns = np.append(intesection_columns, 'year')

intesection_columns = sorted(intesection_columns)

print(len(intesection_columns))
```

```python
# Remove all the columns from df_gdp_developed, df_export_developed, and df_import_developed that are not in intesection_columns

df_gdp_developed = df_gdp_developed[intesection_columns]
df_export_developed = df_export_developed[intesection_columns]
df_import_developed = df_import_developed[intesection_columns]

print(df_gdp_developed.shape)
print(df_export_developed.shape)
print(df_import_developed.shape)

# Place the 'year' column at the beginning of the dataframe
df_gdp_developed = df_gdp_developed[['year'] + [col for col in df_gdp_developed.columns if col != 'year']]
df_export_developed = df_export_developed[['year'] + [col for col in df_export_developed.columns if col != 'year']]
df_import_developed = df_import_developed[['year'] + [col for col in df_import_developed.columns if col != 'year']]
```

# For Developing Countries

## Get GDP Data

```python
df_gdp_developing = pd.read_csv('gdp_developing.csv')
df_gdp_developing['year'] = pd.to_datetime(df_gdp_developing['year'], format='%Y').dt.year
df_gdp_developing = df_gdp_developing[df_gdp_developing['year'] >= 2012]
df_gdp_developing = df_gdp_developing.reset_index(drop=True)

# Drop the columns that contains NaN values
df_gdp_developing = df_gdp_developing.dropna(axis=1)
```

## Get Export Data

```python
df_export_developing = pd.read_csv('export_developing.csv')
df_export_developing['year'] = pd.to_datetime(df_export_developing['year'], format='%Y').dt.year
df_export_developing = df_export_developing[df_export_developing['year'] >= 2012]
df_export_developing = df_export_developing.reset_index(drop=True)

# Drop the columns that contains NaN values
df_export_developing = df_export_developing.dropna(axis=1)
```

## Get Import Data

```python
df_import_developing = pd.read_csv('import_developing.csv')
df_import_developing['year'] = pd.to_datetime(df_import_developing['year'], format='%Y').dt.year
df_import_developing = df_import_developing[df_import_developing['year'] >= 2012]
df_import_developing = df_import_developing.reset_index(drop=True)

# Drop the columns that contains NaN values
df_import_developing = df_import_developing.dropna(axis=1)
```

## Finalizing the data

```python
gdp_developing_columns = df_gdp_developing.columns
export_developing_columns = df_export_developing.columns
import_developing_columns = df_import_developing.columns

intesection_columns = set(gdp_developing_columns).intersection(set(export_developing_columns)).intersection(set(import_developing_columns))

# Remove entry 'year' from the list
intesection_columns.remove('year')

# Convert the intersection columns to a list
intesection_columns = list(intesection_columns)

# Randomly select the 40 strings from the list
intesection_columns = random.sample(intesection_columns, 40)

# If the list not contains 'India' then remove the last entry from the list and append 'India' to the list
if 'India' not in intesection_columns:
    intesection_columns = intesection_columns[:-1]
    intesection_columns = np.append(intesection_columns, 'India')

# Append 'year' to the list
intesection_columns = np.append(intesection_columns, 'year')

intesection_columns = sorted(intesection_columns)

print(len(intesection_columns))
```

```python
# Remove all the columns from df_gdp_developing, df_export_developing, and df_import_developing that are not in intesection_columns

df_gdp_developing = df_gdp_developing[intesection_columns]
df_export_developing = df_export_developing[intesection_columns]
df_import_developing = df_import_developing[intesection_columns]

print(df_gdp_developing.shape)
print(df_export_developing.shape)
print(df_import_developing.shape)

# Place the 'year' column at the beginning of the dataframe
df_gdp_developing = df_gdp_developing[['year'] + [col for col in df_gdp_developing.columns if col != 'year']]
df_export_developing = df_export_developing[['year'] + [col for col in df_export_developing.columns if col != 'year']]
df_import_developing = df_import_developing[['year'] + [col for col in df_import_developing.columns if col != 'year']]
```

# Calculating Trade Openness

```python
# Calculate Trade openness for developed countries

df_trade_openness_developed = pd.DataFrame()

df_trade_openness_developed['year'] = df_gdp_developed['year']

for i in range(0, len(df_gdp_developed.columns)):
    if df_gdp_developed.columns[i] == 'year':
        continue
    df_trade_openness_developed[df_gdp_developed.columns[i]] = (df_export_developed[df_gdp_developed.columns[i]] + df_import_developed[df_gdp_developed.columns[i]]) / df_gdp_developed[df_gdp_developed.columns[i]]

print(df_trade_openness_developed.shape)
```

```python
# Calculate Trade openness for developing countries

df_trade_openness_developing = pd.DataFrame()

df_trade_openness_developing['year'] = df_gdp_developing['year']

for i in range(0, len(df_gdp_developing.columns)):
    if df_gdp_developing.columns[i] == 'year':
        continue
    df_trade_openness_developing[df_gdp_developing.columns[i]] = (df_export_developing[df_gdp_developing.columns[i]] + df_import_developing[df_gdp_developing.columns[i]]) / df_gdp_developing[df_gdp_developing.columns[i]]

print(df_trade_openness_developing.shape)
```

# Calculating Economic Growth

```python
# Caculate the economic growth rate for developed countries as GDP(t) - GDP(t-1) / GDP(t-1) except for the first year

df_economic_growth_rate_developed = pd.DataFrame()

df_economic_growth_rate_developed['year'] = df_gdp_developed['year']

for i in range(0, len(df_gdp_developed.columns)):
    if df_gdp_developed.columns[i] == 'year':
        continue
    df_economic_growth_rate_developed[df_gdp_developed.columns[i]] = df_gdp_developed[df_gdp_developed.columns[i]].pct_change()
    
print(df_economic_growth_rate_developed.shape)
```

```python
# Calculate the economic growth rate for developing countries as GDP(t) - GDP(t-1) / GDP(t-1) except for the first year

df_economic_growth_rate_developing = pd.DataFrame()

df_economic_growth_rate_developing['year'] = df_gdp_developing['year']

for i in range(0, len(df_gdp_developing.columns)):
    if df_gdp_developing.columns[i] == 'year':
        continue
    df_economic_growth_rate_developing[df_gdp_developing.columns[i]] = df_gdp_developing[df_gdp_developing.columns[i]].pct_change()

print(df_economic_growth_rate_developing.shape)
```

# Calculating Correlation

### For Developed Countries

```python
# Calculate the correlation between trade openness and economic growth rate for developed countries except for the first year

correlations = []

for i in range(0, len(df_trade_openness_developed.columns)):
    if df_trade_openness_developed.columns[i] == 'year':
        continue

    # df_correlation_developed[df_trade_openness_developed.columns[i]] = df_trade_openness_developed[df_trade_openness_developed.columns[i]].corr(df_economic_growth_rate_developed[df_trade_openness_developed.columns[i]])
    trade_openness = df_trade_openness_developed[df_trade_openness_developed.columns[i]]
    economic_growth_rate = df_economic_growth_rate_developed[df_trade_openness_developed.columns[i]]

    # Remove the first entry from both trade_openness and economic_growth_rate
    trade_openness = trade_openness[1:]
    economic_growth_rate = economic_growth_rate[1:]

    # Calculate the correlation between trade_openness and economic_growth_rate
    correlations.append(trade_openness.corr(economic_growth_rate))

# Convert the correlations list to a dataframe and store it in df_correlation_developed
df_correlation_developed = pd.DataFrame(correlations)

# Set the rows of df_correlation_developed to the country names
df_correlation_developed.index = df_trade_openness_developed.columns[1:]

# Transpose the dataframe
df_correlation_developed = df_correlation_developed.T

print(df_correlation_developed.size)
```

```python
# Make bar plot of correlation for each country

# Set the figure size
plt.figure(figsize=(200, 60))

# Set the title
plt.title('Correlation between Trade Openness and Economic Growth Rate for Developed Countries', fontsize=70)

# Set the x-axis label
plt.xlabel('Country')

# Set the y-axis label
plt.ylabel('Correlation')

# Plot the bar chart
plt.bar(df_correlation_developed.columns, df_correlation_developed.iloc[0])

# Set the text size of the x-axis ticks
plt.xticks(fontsize=30)

# Set the text size of the y-axis ticks
plt.yticks(fontsize=30)

# Display the plot
plt.show()
```

### Calculating Mean and Standard Deviation for Developed Countries and Developing Countries for each year

```python
# Calulate the mean and standard deviation of for developed countries and developing countries for all the years

years = df_trade_openness_developed['year']
mean_developed = []
standard_deviation_developed = []
mean_developing = []
standard_deviation_developing = []

# trade_open_np_developed = np.array(df_trade_openness_developed)

for i in range(0, len(years)):
    mean = df_trade_openness_developed.iloc[i, 1:].mean()
    mean_developed.append(mean)

    # Calculate standard deviation of trade
    standard_deviation = df_trade_openness_developed.iloc[i, 1:].std()
    standard_deviation_developed.append(standard_deviation)

    mean = df_trade_openness_developing.iloc[i, 1:].mean()
    mean_developing.append(mean)

    # Calculate standard deviation of trade
    standard_deviation = df_trade_openness_developing.iloc[i, 1:].std()
    standard_deviation_developing.append(standard_deviation)

df_statistics = pd.DataFrame({'year': years, 'mean_developed': mean_developed, 'standard_deviation_developed': standard_deviation_developed, 'mean_developing': mean_developing, 'standard_deviation_developing': standard_deviation_developing})

print(df_statistics)
```

# Hypothesis Testing For One Tail

```python
# Calculate the z-score for alpha = 0.05
alpha = 0.05

z_score = stats.norm.ppf(1 - alpha)
print(z_score)
```

```python
n = len(df_trade_openness_developed.columns) - 1

z_calculated = []

for i in range(0, len(df_statistics)):
    z_calculated.append((df_statistics['mean_developed'][i] - df_statistics['mean_developing'][i]) / math.sqrt((df_statistics['standard_deviation_developed'][i] ** 2 / n) + (df_statistics['standard_deviation_developing'][i] ** 2 / n)))

df_statistics['z_calculated'] = z_calculated

df_statistics['Alternate Hypothesis Accepted'] = df_statistics['z_calculated'] > z_score
```

```python
# Calculate the z-score for alpha = 0.05
alpha = 0.05

z_score = stats.norm.ppf(1 - alpha/2)
print(z_score)
```
